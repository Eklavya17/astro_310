```python
from math import *
import matplotlib.pyplot as plot
fctn = input("Enter a function of x: ")
a = float(input("Enter starting value of x: "))
b = float(input("Enter ending value of x: "))
N = int( input("Enter number of samples: "))
def f(x):
    return eval(fctn) #returns the function
I = 0.0
x = a
dx = (b-a)/(N-1)
y = f(a)
# d = 0.0
h = dx
# fn = y + h
ax = []
bx = []

plot.plot(x, y, 'k-', label=r"$f(x)$")
plot.plot(x, I, 'r-', label=r"$\int_0^x f(x')\,dx'$")

print("Step #   x   Int")
z = 0
while x < b:
    xn = x + dx
    yn = f(xn)
    In = I + 0.5*(y+yn)*dx # trapezoidal rule
    plot.plot([x, xn], [y, yn], 'k-')
    plot.plot([x, xn], [I, In], 'r-')
    derivative = (yn - y) / dx
    ax.append(x)
    bx.append(derivative)
    x = xn
    y = yn
    I = In
    if (z % 20 == 0):
        print(z," ",x, "   ", I)
    z+=1
plot.plot(ax,bx, 'b-')
plot.xlabel(r'$x$')
plot.ylabel('Function and integral')
plot.legend()
plot.show()
print("final value of integral = ", I)
# "x**2" -> x^2
# An example of a simple Python program (contd)
```
